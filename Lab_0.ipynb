Let's start by generating white noise. Complete the code below for it.

````python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
````

````python
# Function that returns noise samples
def make_noise( duration=1, sample_rate=8000):
    # YOUR CODE HERE
    return np.random.rand(duration * sample_rate)

# Generate it
duration = 1
sr = 8000
x = make_noise(duration=duration, sample_rate=sr)


# Play the generated sound
sound( x, rate=sr, label='Noise')

plt.plot(x)
plt.show()
````

Generate a sinusoid with a frequency of 440 Hz

````python
# Function that returns a sinusoid
def make_sine( frequency=440, duration=1, sample_rate=8000):
    # YOUR CODE HERE
    t = np.linspace(0, duration, num=int(sample_rate*duration))
    
    return np.sin(2*np.pi*frequency*t)
    

# Generate it
x = make_sine(frequency=440, duration=duration, sample_rate=sr)

# Play the generated sound
sound( x, rate=sr, label='440Hz tone')

plt.plot(x[:80])
plt.show()

# Fourier transform
fft_output = np.fft.rfft(x)
magnitude = [np.sqrt(i.real**2+i.imag**2)/len(fft_output) for i in fft_output]
frequencies = [(i*1.0/len(x)) * sr for i in range(len(x)//2+1)]

plt.xlim(400, 500)
plt.plot(frequencies, magnitude, 'r')
plt.show()
````

Generate a "linear chirp" from 0 Hz to 4 kHz (do not use a chirp function, write the code yourself). What is a chirp you ask? I don't know, google it ...

````python
# Function that returns noise samples
def make_chirp( duration=1, sample_rate=8000):
    # YOUR CODE HERE
    t = np.linspace(0, duration, num=int(sample_rate*duration))
    frequency = 4000 * t
    return np.sin(np.pi*frequency*t)

# Generate it
fs = 8000
x = make_chirp(sample_rate=fs)

# Play the generated sound
sound( x, rate=fs, label='Chirp')

plt.specgram(x, Fs=fs)
plt.show()
````

Generate a sinusoidal tone with an exponentially decreasing amplitude from 100 to 0.0001

````python
# Function that returns noise samples
def make_ping( frequency=440, duration=1, sample_rate=8000):
    # YOUR CODE HERE
    t = np.linspace(0, duration, num=int(sample_rate*duration))
    
    M = sample_rate*duration
    tau = -(M-1) / np.log(0.0001)
    window = signal.exponential(M, 0, tau, False)
    return np.sin(2*np.pi*frequency*t) * window

# Generate it
x = make_ping()

# Play the generated sound
sound( x, rate=sr, label='ping')

plt.plot(x)
plt.show()
````
